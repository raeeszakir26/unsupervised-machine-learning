```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('/content/IRIS.f.csv')
dataset.head(4)
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(10, 10))
sc.dendrogram(sc.linkage(dataset, method='single', metric='euclidean'))
plt.savefig('dendeogram.jpg')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=2, linkage='single')
dataset['predcted']= ac.fit_predict(dataset)
```

```python
dataset
```

```python
sns.pairplot(data=dataset, hue='predcted')
plt.savefig('clustering.jpg')
plt.show()
```
